```python
def Get_value(Variable):
    y = Variable
    Teen_related = [0,0]
    for i in y:
        if i:
            Teen_related[0]+=1
        else:
            Teen_related[1]+=1
    return np.array(Teen_related)
```

```python
Crash_Severity = df['CrashSeverity']
Total_Injured = df['TotalInjured']
Total_Killed = df['TotalKilled']
Youth_Alcohol = df[df['YouthRelated']==True]['AlcoholRelated']
Youth_Drug = df[df['YouthRelated']==True]['DrugRelated']
Youth_DUI = Youth_Alcohol + Youth_Drug
Teen_Alcohol = df[df['TeenRelated']==True]['AlcoholRelated']
Teen_Drug = df[df['TeenRelated']==True]['DrugRelated']
Senior_Alcohol = df[df['SeniorRelated']==True]['AlcoholRelated']
Senior_Drug = df[df['SeniorRelated']==True]['DrugRelated']
```

```python
Crashes_Youth_Alcohol = Get_value(Youth_Alcohol)
Crashes_Youth_Drug = Get_value(Youth_Drug)
Crashes_Youth_DUI = Crashes_Youth_Alcohol + Crashes_Youth_Drug
print(Crashes_Youth_DUI)
```

```python
Crashes_Teen_Alcohol = Get_value(Teen_Alcohol)
Crashes_Teen_Drug = Get_value(Teen_Drug)
Crashes_Teen_DUI = Crashes_Teen_Alcohol + Crashes_Teen_Drug
print(Crashes_Teen_DUI)
```

```python
Crashes_Senior_Alcohol = Get_value(Senior_Alcohol)
Crashes_Senior_Drug = Get_value(Senior_Drug)
Crashes_Senior_DUI = Crashes_Senior_Alcohol + Crashes_Senior_Drug
print(Crashes_Senior_DUI)
```
